```
import pandas as pd
train_dataset=pd.read_csv('iris_training.csv')
test_dataset=pd.read_csv('iris_test.csv')
```

```
features_train=train_dataset.iloc[:,:-1].values
labels_train=train_dataset.iloc[:,-1].values
```

```
features_test=test_dataset.iloc[:,:-1].values
labels_test=test_dataset.iloc[:,-1].values
```

```
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
```

```
classifier=Sequential()
```

```
classifier.add(Dense(units=10, kernel_initializer='uniform',activation='relu',input_dim=4))
```

```
classifier.add(Dense(units=10, kernel_initializer='uniform', activation='relu'))
```

```
classifier.add(Dense(units=3, kernel_initializer='uniform', activation='softmax'))
```

```
classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
encoded_labels_train=np_utils.to_categorical(labels_train)
encoded_labels_test=np_utils.to_categorical(labels_test)
```

```
classifier.fit(features_train, encoded_labels_train, batch_size=10,epochs=500)
```

```
scores=classifier.evaluate(features_test, encoded_labels_test)
```

```
scores
```

```
print("Accuracy: %.2f%%",scores[1]*100)
```
